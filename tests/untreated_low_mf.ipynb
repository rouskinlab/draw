```python
import os, json, sys
sys.path.append(os.path.abspath(os.path.join('..')))
import pandas as pd

from seismic_graph import Study
```

```python
data = []

paths = [
    "/Users/casper/Documents/data/Adam/Astrophage/Astrophage_Data_P3/out/no_DMS__webapp.json"
]

for path in paths:
    with open (path, 'r') as file:
        json_data = json.load(file) 
        data.append(json_data)

study = Study(data=data)
df = study.df
```

```python
df = study.identify_high_mutation_sequences()
```

```python
df
```

```python
# Count the number of sequences with all bases below the threshold vs. those that are not
below_threshold_counts = df['all nt below threshold'].value_counts()
endogenous_mut_counts = df['has endogenous mutation'].value_counts()

print("Sequences with all nucleotides below threshold:", below_threshold_counts.get(True, 0))
print("Sequences with some nucleotides above threshold:", below_threshold_counts.get(False, 0))

print("Sequences with endogenous mutation:", endogenous_mut_counts.get(True, 0))
```
